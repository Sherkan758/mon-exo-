## Tests comparaison et logique 

```python
# 📝 Complétez les fonctions suivantes en utilisant des opérateurs de comparaison et logiques

def est_entre(val: int, borne_min: int, borne_max: int) -> bool:
    """
    Retourne True si val est strictement entre borne_min et borne_max.
    """
    if borne_min < val  < borne_max:
        return True
    else:
        return False

def sont_egaux_ou_negatifs(a: int, b: int) -> bool:
    """
    Retourne True si a et b sont égaux OU si les deux sont négatifs.
    """
    if a == b or (a < 0 and b < 0):
        return True
    else:
        return False

def a_un_age_valide(age: int) -> bool:
    """
    Retourne True si l’âge est entre 18 et 120 inclus.
    """
    if 18 <= age <=120:
        return True
    else:
        return False

def mot_de_passe_valide(mot_de_passe: str) -> bool:
    """
    Retourne True si le mot de passe fait au moins 8 caractères et ne contient pas d’espace.
    """
    if len(mot_de_passe) >= 8 and ' ' not in mot_de_passe:
        return True
    else:
        return False 

def est_non_vide_et_non_null(texte: str) -> bool:
    """
    Retourne True si le texte n’est ni vide ni égal à 'null' ou 'None' (en tant que chaînes).
    """
    if texte and texte.lower() not in ['null', 'none']:
        return True
    else:
        return False
```

```python
# ✅ Cette cellule teste les fonctions définies ci-dessus

def run_logic_tests():
    tests = [
        ("est_entre", est_entre, (5, 1, 10), True),
        ("est_entre", est_entre, (1, 1, 10), False),
        ("est_entre", est_entre, (10, 1, 10), False),

        ("sont_egaux_ou_negatifs", sont_egaux_ou_negatifs, (-2, -2), True),
        ("sont_egaux_ou_negatifs", sont_egaux_ou_negatifs, (-1, -5), True),
        ("sont_egaux_ou_negatifs", sont_egaux_ou_negatifs, (3, 3), True),
        ("sont_egaux_ou_negatifs", sont_egaux_ou_negatifs, (3, -3), False),

        ("a_un_age_valide", a_un_age_valide, (18,), True),
        ("a_un_age_valide", a_un_age_valide, (120,), True),
        ("a_un_age_valide", a_un_age_valide, (17,), False),
        ("a_un_age_valide", a_un_age_valide, (121,), False),

        ("mot_de_passe_valide", mot_de_passe_valide, ("12345678",), True),
        ("mot_de_passe_valide", mot_de_passe_valide, ("1234 678",), False),
        ("mot_de_passe_valide", mot_de_passe_valide, ("short",), False),

        ("est_non_vide_et_non_null", est_non_vide_et_non_null, ("Hello",), True),
        ("est_non_vide_et_non_null", est_non_vide_et_non_null, ("",), False),
        ("est_non_vide_et_non_null", est_non_vide_et_non_null, ("null",), False),
        ("est_non_vide_et_non_null", est_non_vide_et_non_null, ("None",), False),
    ]

    passed = 0
    failed = 0

    for name, func, inputs, expected in tests:
        try:
            result = func(*inputs)
            if result == expected:
                print(f"✅ {name}{inputs} == {expected} → OK")
                passed += 1
            else:
                print(f"❌ {name}{inputs} → {result}, attendu : {expected}")
                failed += 1
        except Exception as e:
            print(f"❌ {name}{inputs} → Erreur inattendue : {e}")
            failed += 1

    print("\n📊 Résumé des tests :")
    print(f"✅ Réussis : {passed}")
    print(f"❌ Échoués : {failed}")

run_logic_tests()
```
